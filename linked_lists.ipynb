```python
class Node:
    def __init__(self,value):
        self.data = value
        self.next = None

def take_user_input():
    values = [int(x) for x in (input().split())]
    head = None
    tail = None
    for val in values:
        newNode = Node(val)
        if head == None:
            head = newNode
            tail = newNode
        else:
            tail.next = newNode
            tail = newNode
    return head 

def print_ll(head):
    temp = head
    while temp != None:
        print(temp.data, end ="->")
        temp = temp.next
    print("None")
        
            
def length_ll(head):
    ans = 0
    temp = head
    while temp != None:
        temp = temp.next
        ans+=1
    return ans
def insert_at_head(head, value):
    newNode = Node(value)

    if head == None:
        head = newNode
    else:
        newNode.next = head
        head = newNode
    return head

def insert_at_tail(head,value):
    newNode = Node(value)
    temp = head
    while temp.next is not None:
        temp = temp.next
    temp.next = newNode
    temp = newNode
    return head

def insert_at_tail_recursively(head,value):
    if head is None:
        newNode = Node(value)
        return newNode
    else:
        head.next = insert_at_tail_recursively(head.next,value)
        return head

def insert_at_index(head,value,index):
    newNode = Node(value) 
    if index == 0:
        newNode.next = head
        head = newNode
    count = 0
    temp = head
    while count < index-1 and temp.next != None :
        temp = temp.next
        count = count+1

    newNode.next = temp.next
    temp.next = newNode 
    return head
        
```

```python
def merge_two_sorted(llst1, llst2):
    if llst1 is None:
        return llst2
    if llst2 is None:
        return llst1

    finallst_head = None
    finallst_tail = None

    while llst1 is not None and llst2 is not None:
        if llst1.data < llst2.data:
            if finallst_head is None:
                finallst_head = llst1
                finallst_tail = llst1
            else:
                finallst_tail.next = llst1
                finallst_tail = llst1
            llst1 = llst1.next
        else:
            if finallst_head is None:
                finallst_head = llst2
                finallst_tail = llst2
            else:
                finallst_tail.next = llst2
                finallst_tail = llst2
            llst2 = llst2.next

    # Append any remaining nodes
    if llst1 is not None:
        finallst_tail.next = llst1
    if llst2 is not None:
        finallst_tail.next = llst2

    return finallst_head
```
